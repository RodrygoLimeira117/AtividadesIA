```python
# PRÉ-PROCESSAMENTO COM O DATASET IRIS

# Recebemos uma amostra de dados de flores Iris,
# mas parece que o equipamento de medição estava com defeito.
# Sua missão é limpar e preparar este dataset para que ele possa ser usado
# em uma análise futura. Siga os passos e preencha o código onde for indicado.

# 0. Importando as Bibliotecas Necessárias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Criando um dataset 'iris_alterado.csv' simulado para o exercício
# (Este passo é apenas para que o código seja executável)
from sklearn.datasets import load_iris
iris = load_iris()
df_original = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                           columns= iris['feature_names'] + ['target'])
df_original.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
species_map = {0.0: 'Iris-setosa', 1.0: 'Iris-versicolor', 2.0: 'Iris-virginica'}
df_original['Species'] = df_original['Species'].map(species_map)
# Introduzindo erros no dataset
df_alterado = df_original.copy()
for col in ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']:
    df_alterado.loc[df_alterado.sample(frac=0.1).index, col] = np.nan
df_alterado.loc[[10, 20, 30], 'SepalWidthCm'] = 5.0 # Outliers
df_alterado.to_csv('iris_alterado.csv', index=False)


# Configurando o estilo dos gráficos para uma melhor visualização
sns.set_style('whitegrid')

# 1. Carregando e Explorando os Dados
print("--- Passo 1: Investigando a Cena do Crime (os Dados) ---")

# Tarefa 1.1: Carregue o dataset 'iris_alterado.csv' para um DataFrame.
# Dica: Use a função pd.read_csv()
df = pd.read_csv('iris_alterado.csv')

# Tarefa 1.2: Visualize as 5 primeiras linhas para ter uma ideia inicial.
print("As 5 primeiras amostras são:")
print(df.head())

# Tarefa 1.3: Use o método .info() para obter um resumo técnico.
# Preste atenção nos tipos de dados e na contagem de valores não-nulos.
print("\nRelatório técnico do dataset:")
df.info()

# Tarefa 1.4: Use .isnull().sum() para contar exatamente quantos valores
# faltantes existem em cada coluna.
print("\nContagem de registros corrompidos (dados faltantes):")
print(df.isnull().sum())

# 2. Tratando os Dados Faltantes
print("\n--- Passo 2: Consertando as Medidas Faltantes ---")
# Vimos que várias colunas numéricas têm dados faltando.
# Uma estratégia comum é preencher esses buracos com a MÉDIA ou a MEDIANA da coluna.
# A mediana é geralmente mais segura se houver outliers. Vamos usá-la.

# Tarefa 2.1: Calcule a mediana para cada uma das 4 colunas de medidas.
# Dica: use o método .median()
median_sepal_length = df['SepalLengthCm'].median()
median_sepal_width = df['SepalWidthCm'].median()
median_petal_length = df['PetalLengthCm'].median()
median_petal_width = df['PetalWidthCm'].median()

print(f"\nMediana de SepalLengthCm: {median_sepal_length}")
print(f"Mediana de SepalWidthCm: {median_sepal_width}")
print(f"Mediana de PetalLengthCm: {median_petal_length}")
print(f"Mediana de PetalWidthCm: {median_petal_width}")

# Tarefa 2.2: Preencha os valores faltantes de cada coluna com sua respectiva mediana.
# Dica: use o método .fillna()
df['SepalLengthCm'].fillna(median_sepal_length, inplace=True)
df['SepalWidthCm'].fillna(median_sepal_width, inplace=True)
df['PetalLengthCm'].fillna(median_petal_length, inplace=True)
df['PetalWidthCm'].fillna(median_petal_width, inplace=True)

# Tarefa 2.3: Verifique se ainda existem dados faltantes.
print("\nVerificação após o primeiro conserto:")
print(df.isnull().sum())

# 3. Identificando e Tratando Outliers
print("\n--- Passo 3: Encontrando Outliers ---")

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Boxplots das Características da Flor Iris - ANTES do Tratamento')

sns.boxplot(ax=axes[0, 0], x=df['SepalLengthCm'])
axes[0, 0].set_title('Sepal Length')
sns.boxplot(ax=axes[0, 1], x=df['SepalWidthCm'])
axes[0, 1].set_title('Sepal Width')
sns.boxplot(ax=axes[1, 0], x=df['PetalLengthCm'])
axes[1, 0].set_title('Petal Length')
sns.boxplot(ax=axes[1, 1], x=df['PetalWidthCm'])
axes[1, 1].set_title('Petal Width')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# DISCUSSÃO: Analisando os gráficos, quais colunas parecem ter valores
# que se desviam muito do padrão (outliers)?
# Resposta: A coluna 'SepalWidthCm' é a que apresenta os outliers mais evidentes.

# Tarefa 3.1: Vamos tratar os outliers da coluna 'SepalWidthCm' usando o método IQR.
# Calcule o Q1, Q3, e o IQR para a coluna 'SepalWidthCm'.
Q1 = df['SepalWidthCm'].quantile(0.25)
Q3 = df['SepalWidthCm'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR

print(f"\nAnálise de 'SepalWidthCm':")
print(f"Q1: {Q1:.2f}")
print(f"Q3: {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Limite superior para outliers: {limite_superior:.2f}")

# Tarefa 3.2: Substitua qualquer valor em 'SepalWidthCm' que seja MAIOR que o
# limite_superior pelo próprio valor do limite_superior (técnica de "capping").
# Dica: Use np.where(condição, valor_se_verdadeiro, valor_se_falso)
df['SepalWidthCm'] = np.where(df['SepalWidthCm'] > limite_superior, limite_superior, df['SepalWidthCm'])

print("\nOutliers em 'SepalWidthCm' tratados com sucesso.")

# 4. Verificação Final
print("\n--- Passo 4: Inspeção Final ---")

# Tarefa 4.1: Gere novamente o boxplot para 'SepalWidthCm' para ver o resultado.
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['SepalWidthCm'])
plt.title('Boxplot de SepalWidthCm - APÓS o Tratamento')
plt.show()

# Tarefa 4.2: Use o método .describe() para ver as estatísticas finais do dataset.
print("\nEstatísticas descritivas do dataset limpo:")
print(df.describe())
```
